### Extracting Table

```python
url = 'https://en.wikipedia.org/wiki/World_population'

response = requests.get(url).content

soup = BeautifulSoup(response, 'lxml')
soup.prettify()
```

```python
table = soup.find('table', class_="wikitable floatright")
table
```

```python
table_data = []
for row in table.find_all('tr')[1:-1]:  # excluding first and last rows
    columns = row.find_all('td')
    table_data.append([column.text for column in columns[1:]])  #[:1]selecting columns of the table from 1st index instead 0th. 
    
table_data
```

### Convert the above format to a dataframe

```python

import pandas as pd
WorldPop= pd.DataFrame(table_data, columns=['Countries', '2000', '2015', '2030'])
WorldPop
```

```python
WorldPop['2030'] = WorldPop['2030'].str.replace("\n", "")
WorldPop
```
